```python
def clean_name(x):
    return (str(x)
            .strip()
            .title()
            .replace(" ", "")
            .replace("-", "")
            .replace("District", "")
            .replace("Dt", "")
            .replace(".", "")
            )

crop['State_Name'] = crop['State_Name'].apply(clean_name)
crop['District_Name'] = crop['District_Name'].apply(clean_name)
weather['State_Name'] = weather['State_Name'].apply(clean_name)
weather['District_Name'] = weather['District_Name'].apply(clean_name)
```

```python
import re

def advanced_clean_name(x):
    x = str(x)
    x = re.sub(r'[^A-Za-z]', '', x)  # remove all non-letters
    x = x.title()
    x = x.replace("District", "").replace("Dt", "").replace("Dist", "")
    return x.strip()

crop['District_Name'] = crop['District_Name'].apply(advanced_clean_name)
weather['District_Name'] = weather['District_Name'].apply(advanced_clean_name)
```

```python
crop_weather = pd.merge(crop, weather, on=['State_Name', 'District_Name'], how='left')
```

```python
crop_weather.isnull().sum()
```

```python
missing_by_state = crop_weather[crop_weather['Rainfall_Rabi'].isnull()].groupby('State_Name')['District_Name'].nunique().reset_index()
missing_by_state.columns = ['State_Name', 'Missing_Districts']
print(missing_by_state.sort_values('Missing_Districts', ascending=False).head(10))
```
