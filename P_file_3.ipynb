```python
import pandas as pd
import numpy as np
import os
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
CW=pd.read_csv('data/process/crop_weather_cleaned_M1.csv')
```

```python
soil=pd.read_csv("data/process/soil_clean_N1.csv")
```

```python
fertilizer=pd.read_csv("data/process/Fertilizer_clean_N1.csv")
```

```python
CW_1=CW.copy()
```

```python
CW_1
```

```python
def clean_text(x):
    return str(x).strip().title()

for df in [CW_1, soil, fertilizer]:
    if 'State_Name' in df.columns:
        df['State_Name'] = df['State_Name'].apply(clean_text)
    if 'District_Name' in df.columns:
        df['District_Name'] = df['District_Name'].apply(clean_text)
```

```python
merged1 = pd.merge(CW_1, soil, on='District_Name', how='left')
```

```python
merged1.head()
```

```python
merged1.shape
```

```python
merged1.isnull().sum()
```

```python
final_data = pd.merge(merged1, fertilizer, on='State_Name', how='left')
```

```python
soil_cols = ['Zn', 'Fe', 'Cu', 'Mn','B','S']
for col in soil_cols:
    final_data[col] = final_data.groupby('State_Name')[col].transform(lambda x: x.fillna(x.mean()))
```

```python
final_data.isnull().sum()
```

```python
final_data_copy=final_data.copy()
```

```python
final_data.head()
```

```python
final_data_copy.head()
```

```python
def clean_state(x):
    return str(x).strip().title()

fertilizer['State_Name'] = fertilizer['State_Name'].apply(clean_state)
final_data['State_Name'] = final_data['State_Name'].apply(clean_state)
```

```python
fertilizer_cleaned = fertilizer.groupby('State_Name', as_index=False).mean(numeric_only=True)
```

```python
print(fertilizer_cleaned['State_Name'].nunique(), fertilizer_cleaned.shape[0])
```

```python
fert_cols = fertilizer_cleaned.columns.drop('State_Name')

for col in fert_cols:
    final_data[col] = final_data['State_Name'].map(fertilizer_cleaned.set_index('State_Name')[col])
```

```python
final_data.isnull().sum()
```

```python
final_master = final_data.dropna()
```
