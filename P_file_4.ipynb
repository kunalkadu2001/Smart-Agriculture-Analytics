```python
yearly_data = (
    final_data.groupby(['Crop_Year', 'Crop'])['Production']
    .sum()
    .reset_index()
)

years = sorted(final_data['Crop_Year'].unique())

for year in years:
    temp = yearly_data[yearly_data['Crop_Year'] == year].nlargest(5, 'Production')
    
    plt.figure(figsize=(10, 6))
    sns.barplot(data=temp, x='Production', y='Crop', palette='viridis')
    plt.title(f"Top 5 Crops by Production in {year}", fontsize=14, weight='bold')
    plt.xlabel("Total Production")
    plt.ylabel("Crop Name")
    plt.tight_layout()
    plt.show()
```

```python
selected_crops = ['Coconut', 'Sugarcane']
season_crop_prod = (
    final_data[final_data['Crop'].isin(selected_crops)]
    .groupby(['Season', 'Crop'])['Production']
    .sum()
    .reset_index()
)
season_crop_prod = season_crop_prod.sort_values('Production', ascending=False)

# Plot
plt.figure(figsize=(10, 6))
sns.barplot(
    data=season_crop_prod,
    x='Season',
    y='Production',
    hue='Crop',
    palette='viridis'
)
plt.title("Comparison of Coconut & Sugarcane Production Across Seasons", fontsize=14, weight='bold')
plt.xlabel("Season")
plt.ylabel("Total Production")
plt.legend(title='Crop', loc='upper right')
plt.tight_layout()
plt.show()
```

```python
rainfall_by_state = (
    final_data.groupby('State_Name')[['Rainfall_Rabi', 'Rainfall_Kharif']]
    .mean()
    .reset_index()
)
top5_rabi = rainfall_by_state.nlargest(5, 'Rainfall_Rabi')
top5_kharif = rainfall_by_state.nlargest(5, 'Rainfall_Kharif')

plt.figure(figsize=(8, 5))
sns.barplot(data=top5_rabi, x='Rainfall_Rabi', y='State_Name', palette='Blues_r')
plt.title("Top 5 States with Highest Rainfall in Rabi Season", fontsize=14, weight='bold')
plt.xlabel("Average Rainfall (mm)")
plt.ylabel("State")
plt.tight_layout()
plt.show()

plt.figure(figsize=(8, 5))
sns.barplot(data=top5_kharif, x='Rainfall_Kharif', y='State_Name', palette='Greens_r')
plt.title("Top 5 States with Highest Rainfall in Kharif Season", fontsize=14, weight='bold')
plt.xlabel("Average Rainfall (mm)")
plt.ylabel("State")
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import os
```

```python
final_data_copy2=final_data.copy()
```

```python
fy_cols = ['fy_2014_15','fy_2015_16','fy_2016_17','fy_2017_18',
           'fy_2018_19','fy_2019_20','fy_2020_21','fy_2021_22',
           'fy_2022_23','fy_2023_24','fy_2024_25']
df_long = final_data_copy2.melt(id_vars=['fertiliser_type'], value_vars=fy_cols,
                       var_name='Financial_Year', value_name='Production_LMT')

df_long['Financial_Year'] = df_long['Financial_Year'].str.replace('fy_', '').str.replace('_', '-')

plt.figure(figsize=(12,6))
sns.lineplot(data=df_long, x='Financial_Year', y='Production_LMT', hue='fertiliser_type', marker='o')
plt.title("Fertilizer Production Trend Over Years", fontsize=14, weight='bold')
plt.xlabel("Financial Year")
plt.ylabel("Production (LMT)")
plt.xticks(rotation=45)
plt.legend(title='Fertilizer Type')
plt.tight_layout()
plt.show()
```

```python
fy_cols = [col for col in final_data.columns if 'fy_' in col]  # detect all FY columns
final_data['Total_LMT'] = final_data[fy_cols].sum(axis=1)
```

```python
final_data.head()
```

```python
sector_prod = final_data.groupby('sector')['Total_LMT'].sum().reset_index()
plt.figure(figsize=(7,5))
sns.barplot(data=sector_prod.sort_values('Total_LMT', ascending=False),
            x='sector', y='Total_LMT', palette='viridis')
plt.title("Sector-wise Total Fertilizer Production (LMT)", fontsize=14, weight='bold')
plt.ylabel("Total LMT")
plt.xlabel("Sector")
plt.tight_layout()
plt.show()
```

```python
final_data.to_csv("data/process/final_master.csv",index=False)
```

```python
state_prod = final_data.groupby('State_Name')['Total_LMT'].sum().reset_index()
top_states = state_prod.sort_values('Total_LMT', ascending=False).head(10)
plt.figure(figsize=(8,6))
sns.barplot(data=top_states, x='Total_LMT', y='State_Name', palette='crest')
plt.title("Top 10 States by Fertilizer Production (LMT)", fontsize=14, weight='bold')
plt.xlabel("Total LMT")
plt.ylabel("State")
plt.tight_layout()
plt.show()
```

```python
fert_type_prod = final_data.groupby('fertiliser_type')['Total_LMT'].sum().reset_index()
plt.figure(figsize=(8,5))
sns.barplot(data=fert_type_prod.sort_values('Total_LMT', ascending=False),
            x='fertiliser_type', y='Total_LMT', palette='magma')
plt.title("Total Production by Fertilizer Type (LMT)", fontsize=14, weight='bold')
plt.xlabel("Fertilizer Type")
plt.ylabel("Total LMT")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
company_prod = final_data.groupby('company')['Total_LMT'].sum().reset_index()
top_companies = company_prod.sort_values('Total_LMT', ascending=False).head(10)
plt.figure(figsize=(10,6))
sns.barplot(data=top_companies, x='Total_LMT', y='company', palette='coolwarm')
plt.title("Top 10 Companies by Fertilizer Production (LMT)", fontsize=14, weight='bold')
plt.xlabel("Total LMT")
plt.ylabel("Company")
plt.tight_layout()
plt.show()
```
